```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
question_1a = pd.read_csv('../data/Question_1a_for_visual.csv')

question_1a
```

```python
plt.plot(question_1a['brand'], question_1a['dist_rcpt_cnt'])
plt.title('Top 5 (10) Brands by Receipts Scanned - 21yr+')
plt.legend(["Distinct Receipt Count"])
plt.xticks(rotation=45, ha = 'right')
plt.yticks(np.arange(int(min(question_1a['dist_rcpt_cnt'])), int(max(question_1a['dist_rcpt_cnt'])) + 1, step=1))
plt.xlabel('Brands')
plt.ylabel('Distinct Receipt Count')
plt.savefig('top_5_brands_by_receipts_scanned_21up.png', bbox_inches = 'tight')
```

```python
question_2a = pd.read_csv('../data/Question_2a_for_visual.csv')

question_2a
```

```python
plt.bar(question_2a['brand'], question_2a['sum_final_sale'])
plt.title('Top 5 (10) Brands by Sales - 6month+ Account Lifespan')
plt.legend(["Sum Final Sales"])
plt.xticks(rotation=45, ha = 'right')
plt.xlabel('Brands')
plt.ylabel('Sum Final Sales (in Dollars)')
plt.savefig('top_5_brands_by_sum_sales_6monthsup.png', bbox_inches = 'tight')
```
